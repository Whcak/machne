```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
from scipy.stats import  norm
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline  

data_train = pd.read_csv(r'E:/谷歌浏览器/house-prices-advanced-regression-techniques/train.csv',engine='python')
data_train
#读取数据
```

```python
total = data_train.isnull().sum().sort_values(ascending=False)
precent = (data_train.isnull().sum()/data_train.isnull().count()).sort_values(ascending=False)
miss_train = pd.concat([total,precent],axis=1,keys=['Total','Precent'])
miss_train.head(20)
#缺失值统计
```

```python
data_train = data_train.drop(miss_train[miss_train['Precent']>.15].index, axis=1)
#出去缺失率大于15%的列
```

```python
garage = ['GarageType','GarageFinish','GarageQual','GarageCond']
for i in garage:
    data_train[i].fillna("Missing",inplace=True)
    #用Missing来补充没有车库而造成的与车库相关属性的缺失
```

```python
data_train[garage].isnull().any(axis=0)
```

```python
data_train['GarageType'].unique()
```

```python
data_train["GarageYrBlt"].fillna(1850,inplace=True)
#对车库建造缺失的数据用1850填充，表面车库是年久的
```

```python
bsmt = ['BsmtExposure','BsmtFinType1','BsmtFinType2','BsmtCond','BsmtQual']
for i in bsmt:
    data_train[i].fillna("Missing",inplace=True)
    #用Missing来填充没有地下室而造成与地下室相关属性的缺失
```

```python
data_train["MasVnrArea"].fillna(0,inplace=True)
#用0填充没装修过的面积
```

```python
data_train["MasVnrType"].fillna("Missing",inplace=True)
#用Missing填充没装修过的房子
```

```python
data_train["Electrical"].fillna(data_train["Electrical"].mode().values[0],inplace=True)
#用众数填充缺失值很少的属性
```

```python
total = data_train.isnull().sum().sort_values(ascending=False)
precent = (data_train.isnull().sum()/data_train.isnull().count()).sort_values(ascending=False)
miss_train = pd.concat([total,precent],axis=1,keys=['Total','Precent'])
miss_train.head(20)
#统计是否还有缺失值
```

```python
from sklearn import preprocessing
from sklearn import linear_model, svm, gaussian_process
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import numpy as np

# 获取数据
cols = ['OverallQual','GrLivArea', 'GarageCars','TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt']
x = data_train[cols].values
y = data_train['SalePrice'].values
x_scaled = preprocessing.StandardScaler().fit_transform(x)
y_scaled = preprocessing.StandardScaler().fit_transform(y.reshape(-1,1))
X_train,X_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.33, random_state=42)

modeals = {
        'svm':svm.SVR(), 
        'RandomForestRegressor':RandomForestRegressor(n_estimators=400),
        'BayesianRidge':linear_model.BayesianRidge()
       }
for modeal in modeals:
    try:
        modeals[modeal].fit(X_train, y_train)
        y_pred = modeals[modeal].predict(X_test)
        print(modeal + " cost:" + str(np.sum(y_pred-y_test)/len(y_pred)) )
    except Exception as e:
        print(modeal + " Error:")
        print(str(e))
```
